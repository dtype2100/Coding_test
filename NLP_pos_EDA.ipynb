```python
mecab_pos = []
for i in tqdm(train_df['overview']):
  mecab = Mecab()
  tags = mecab.pos(i)
  mecab_pos.append(tags)
```

```python
okt_pos = []
for i in tqdm(train_df['overview']):
  okt = Okt()
  okt_tags = okt.pos(i)
  okt_pos.append(okt_tags) 
```

```python
word = []
tag = []
all_words = []
all_tags = []
for i, j in mecab_pos[0]:
  print(i, j)
  all_words.append(i)
  all_tags.append(j)
  if j == 'NNG':
    tag.append(j)
    word.append(i)
```
